**Import Libraries**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import os
```

**Create / Load Data Storage**

```python
file_name = "tasks.csv"

if not os.path.exists(file_name):
    df = pd.DataFrame(columns=["Task", "Difficulty", "Time", "Priority"])
    df.to_csv(file_name, index=False)

def load_tasks():
    return pd.read_csv(file_name)

def save_tasks(df):
    df.to_csv(file_name, index=False)
```

**Add Task Function**

```python
def add_task():
    df = load_tasks()

    task = input("Enter Task Name: ")
    difficulty = int(input("Difficulty (1-5): "))
    time = int(input("Estimated Time (minutes): "))
    priority = int(input("Priority (1=Low, 2=Medium, 3=High): "))

    new_task = pd.DataFrame({
        "Task": [task],
        "Difficulty": [difficulty],
        "Time": [time],
        "Priority": [priority]
    })

    df = pd.concat([df, new_task], ignore_index=True)
    save_tasks(df)

    print("Task Added Successfully")
```

**Remove Task**

```python
def remove_task():
    df = load_tasks()

    task_name = input("Enter Task Name to Remove: ")

    df = df[df["Task"] != task_name]
    save_tasks(df)

    print("Task Removed")
```

**List Tasks**

```python
def list_tasks():
    df = load_tasks()

    if df.empty:
        print("No tasks available")
    else:
        print(df)
```

**ML Recommendation System**

```python
def recommend_task():
    df = load_tasks()

    if len(df) < 2:
        print("Not enough data for ML recommendation")
        return

    X = df[["Difficulty", "Time"]]
    y = df["Priority"]

    model = DecisionTreeClassifier()
    model.fit(X, y)

    difficulty = int(input("Enter Difficulty: "))
    time = int(input("Enter Time: "))

    prediction = model.predict([[difficulty, time]])

    print("Recommended Priority:", prediction[0])
```

**Main Program**

```python
def main():

    while True:

        print("\nTask Manager")
        print("1 Add Task")
        print("2 Remove Task")
        print("3 List Tasks")
        print("4 Recommend Priority")
        print("5 Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_task()

        elif choice == "2":
            remove_task()

        elif choice == "3":
            list_tasks()

        elif choice == "4":
            recommend_task()

        elif choice == "5":
            break

        else:
            print("Invalid choice")
```

**Run the Program**

```python
main()
```
